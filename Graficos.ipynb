# Importar Librerías 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importar Datos

```python
df = pd.read_csv('.\\BikePrices.csv')
df.head()
```

# Gráficos

```python
#Selling Price y owner

plt.figure()

plt.bar(df["Owner"],df["Selling_Price"], color = "tab:green")
plt.ylabel("Precio de venta")
plt.xlabel("# de Dueños")
plt.title("Precio de venta según el # de dueños")
plt.show()
```

```python
#Tipo de venta

plt.figure()
tipo = df["Seller_Type"].value_counts()
plt.figure(figsize=(5, 5))
# Graficar el pastel sin etiquetas externas
wedges, texts, autotexts = plt.pie(tipo, autopct='%1.1f%%', startangle=140)
# Agregar la leyenda con las categorías
plt.legend(wedges, tipo.index, title="Tipo de Venta", loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Tipo de venta")
plt.show()
```

```python
#Precio de motos promedio con el tiempo
import seaborn as sns
precioA = df.groupby("Year")["Ex_Showroom_Price"].mean()
plt.figure(figsize=(10,5))
sns.lineplot(x=precioA.index, y=precioA.values, marker="o", linestyle="-", color="b")
plt.xlabel("Año de Fabricación")
plt.ylabel("Precio Promedio en Showroom")
plt.title("Precio de Motos con el Tiempo")
plt.grid(True)

plt.show()
```
